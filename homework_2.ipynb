```python
import mlflow
mlflow.set_tracking_uri("http://localhost:5000")
```

```python
ordered_runs = mlflow.search_runs(search_all_experiments=True, order_by=["metrics.rmse ASC"])
ordered_runs
```

```python
best_run_id = ordered_runs.iloc[0]["run_id"]
best_run_tag = ordered_runs.iloc[0]["tags.mlflow.runName"]
```

```python
from mlflow.exceptions import MlflowException
from mlflow.tracking import MlflowClient
client = MlflowClient(tracking_uri="http://localhost:5000")
try:
    for rm in client.search_registered_models():
        print(dict(rm), indent=4)
except MlflowException as e:
    print("It's not possible to access the model registry.")
```

```python
reg_model_name = "random_forest_regression"
signature = mlflow.models.infer_signature(best_run_tag, ordered_runs.iloc[0]["metrics.rmse"])
mlflow.sklearn.log_model(best_run_id, "sk_learn", serialization_format="cloudpickle", registered_model_name=reg_model_name, signature=signature)
```
